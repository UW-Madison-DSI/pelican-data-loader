# Accessing files on Pelicanfs

- pelicanfs automatically handle the caching

```python
import os

os.chdir("..")
```

```python
from pelicanfs.core import PelicanFileSystem
import pandas as pd
```

```python
# pelfs = PelicanFileSystem("pelican://osg-htc.org")
pelfs = PelicanFileSystem("pelican://uwdf-director.chtc.wisc.edu")
```

```python
pelfs.ls("/wisc.edu/dsi/pytorch")
```

```python
csv_file = "/wisc.edu/dsi/pytorch/bird_migration_data.csv"
```

### Open with pelfs API

```python
with pelfs.open(csv_file, mode="r") as f:
    df = pd.read_csv(f)
df.head()
```

### Open directly from pandas

```python
df = pd.read_csv("pelican://uwdf-director.chtc.wisc.edu/wisc.edu/dsi/pytorch/bird_migration_data.csv")
df.head()
```
